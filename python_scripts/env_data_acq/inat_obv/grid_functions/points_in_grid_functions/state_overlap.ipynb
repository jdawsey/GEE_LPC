```python
folder_path = 'D:/tifs/broad_xgb_classification'
# Load the shapefiles
points_gdf = gpd.read_file(f'{folder_path}/state_shrub_shps/new_mexico_shrubs.shp')
states_gdf = gpd.read_file(f'{folder_path}/states_shp/US_States.shp')

# Ensure both GeoDataFrames have the same CRS
if points_gdf.crs != states_gdf.crs:
    states_gdf = states_gdf.to_crs(points_gdf.crs)

# Perform the spatial join
points_with_states = gpd.sjoin(points_gdf, states_gdf, how='left', predicate='within')

# Display the resulting GeoDataFrame
points_with_states.head()
```

```python
extracted_col = points_with_states["STATE_NAME"]
points_gdf['state'] = points_with_states["STATE_NAME"]
points_gdf.head()
```
