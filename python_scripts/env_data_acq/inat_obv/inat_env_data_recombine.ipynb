```python
import os
import pandas as pd
```

```python
# Define the folder path containing the CSV files
#folder_path = f'D:/!!Research/rgee_test/GEE_LPC/point_data/mesq_location_predictions/poly{poly_num}/env_batch_csv'
folder_path = f'C:/Users/Justin-Laptop/Documents/Research/GEE_LPC/point_data/inat_mesq_obsv/inat_points_env_data/env_batch_csv'

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through each file in the folder
for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        # Read the CSV file into a DataFrame
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# Display or do further processing with the merged DataFrame
merged_df.head()
```

```python
# Export the merged DataFrame to a CSV file
#output_csv_path = f'D:/!!Research/rgee_test/GEE_LPC/point_data/mesq_location_predictions/poly{poly_num}/poly{poly_num}_env_data.csv'
output_csv_path = f'C:/Users/Justin-Laptop/Documents/Research/GEE_LPC/point_data/inat_mesq_obsv/inat_points_env_data/inat_points_env_data.csv'
merged_df.to_csv(output_csv_path, index=False)

print("Merged DataFrame has been exported to:", output_csv_path)
```
